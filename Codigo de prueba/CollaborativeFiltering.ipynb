```python

juegos = pd.read_csv('../data/bgg-limpio-reviews.csv')
more_info = pd.read_csv('../data/games_detailed_info.csv')
juegos.columns = juegos.columns.str.lower()
more_info.columns = more_info.columns.str.lower()
more_info_f = more_info[['id', 'description', 'yearpublished', 'minplayers', 'maxplayers', 'playingtime', 'boardgamecategory', 'boardgamemechanic', 'boardgamedesigner']]
juegos.head(4)
more_info_f.head(4)
```

```python
juegos_merged_f = juegos.merge(more_info_f, on='id', how='inner')
juegos.dropna(inplace=True)
```

```python
juegos_merged_f.head(3)
```

```python
juegos_merged_f = juegos_merged_f.astype(str)
```

```python
sample_df = juegos_merged_f.iloc[:3000].copy()

# sample_df['info_gameboard'] = sample_df['description'] + " " + sample_df['yearpublished']   
# sample_df.head(3)

sample_df['info_gameboard'] = sample_df[
    ['description', 'yearpublished', 'minplayers', 'maxplayers', 
     'playingtime', 'boardgamecategory', 'boardgamemechanic', 'boardgamedesigner']
].astype(str).agg(' '.join, axis=1)



# tabla_juegos = juegos_merged_f.drop(columns=['description', 'yearpublished', 'minplayers', 'maxplayers', 'playingtime', 'boardgamecategory', 'boardgamemechanic', 'boardgamedesigner'])
# juegos_merged_f.head(2)
```

```python
sample_df = sample_df.drop(columns=['description', 'yearpublished', 'minplayers', 'maxplayers', 'playingtime', 'boardgamecategory', 
                                             'boardgamemechanic', 'boardgamedesigner'])
sample_df.head(5)
```

```python
print(type(sample_df))
print(sample_df.head())


sample_df.dropna(axis=0, inplace=True)
sample_df.head(3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')

vector = cv.fit_transform(sample_df['info_gameboard']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)

similarity
```

```python
valor = sample_df[sample_df['name'] == 'CATAN'].index[0]
print(int(valor))
```

```python
def recomendar(juego):
    index = sample_df[sample_df['name'] == juego].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    
    for i in distances[1:6]:
        recomendado = sample_df.iloc[i[0]]
        print(recomendado['name'], "-", recomendado['info_gameboard'][:200])
```

```python
recomendar('CATAN')
```
